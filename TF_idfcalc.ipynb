```python
import glob2

filenames = glob2.glob('/content/data/*.txt')  # list of all .txt files in the directory

with open('outfile.txt', 'w') as f:
    for file in filenames:
        with open(file) as infile:
            f.write(infile.read()+'\n')

files= open('outfile.txt').read().replace('\n', '').replace('\\','')

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
doc_1=files
```

```python
response=tfidf.fit_transform([doc_1])
```

```python
print(len(tfidf.vocabulary_))
```
